```python
import mysql.connector
mydb = mysql.connector.connect(host='localhost', database='dm_etl1', user='root', password='Csoares#12')
df = pd.read_sql('select a.*, sum(b.TotalPrice), count(b.SalesID) from custdetails a, salesdetails b where a.customerID = b.CustomerID group by a.CustomerID', con=mydb)
```

```python
df.rename(columns = {'sum(b.TotalPrice)':'Sum_TotalPrice'}, inplace = True)
df.rename(columns = {'count(b.SalesID)':'Count_TotalSales'}, inplace = True)
```

```python
bins = [0, 99999, 99999999, 9999999999]
labels = ['LOW','MID','HIGH']
df['Customer_Category'] = pd.cut(df['Sum_TotalPrice'], bins=bins, 
                                       labels=labels, right=True, include_lowest=True)
df
```

```python
#Slicing of data to devide it into target and feature set
X=df[['Sum_TotalPrice','Count_TotalSales']]
y=df[['Customer_Category']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.25, random_state = 18)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_tree = RandomForestClassifier(random_state=18, n_estimators=100)
model_tree.fit(X_train, y_train)
```

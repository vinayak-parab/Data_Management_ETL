# Split the dataframe into multiple dataframes 

creation of Customer data frame

```python
# perform group by operation to get the counts 
# tmpcstdf = df.groupby(['CustomerID','CustFirstName','CustMiddleInitial','CustLastName'])['SalesID'].nunique()
tmpcstname = df.groupby(['CustomerID'])['CustFirstName'].value_counts(dropna = False)
tmpcstname=tmpcstname.to_frame()
tmpcstname.columns=['counts']
tmpcstname=tmpcstname.reset_index(level=['CustFirstName', 'CustomerID'])
tmpcstname=tmpcstname.drop_duplicates(subset='CustomerID', keep='first')
tmpcstname
```

```python
# perform group by operation to get the counts 
# tmpcstdf = df.groupby(['CustomerID','CustFirstName','CustMiddleInitial','CustLastName'])['SalesID'].nunique()
tmpcstMid = df.groupby(['CustomerID'])['CustMiddleInitial'].value_counts(dropna = False)
tmpcstMid=tmpcstMid.to_frame()
tmpcstMid.columns=['counts']
tmpcstMid=tmpcstMid.reset_index(level=['CustMiddleInitial', 'CustomerID'])
tmpcstMid=tmpcstMid.drop_duplicates(subset='CustomerID', keep='first')
tmpcstMid
```

```python
# perform group by operation to get the counts 
# tmpcstdf = df.groupby(['CustomerID','CustFirstName','CustMiddleInitial','CustLastName'])['SalesID'].nunique()
tmpcstlast = df.groupby(['CustomerID'])['CustLastName'].value_counts(dropna = False)
tmpcstlast=tmpcstlast.to_frame()
tmpcstlast.columns=['counts']
tmpcstlast=tmpcstlast.reset_index(level=['CustLastName', 'CustomerID'])
tmpcstlast=tmpcstlast.drop_duplicates(subset='CustomerID', keep='first')
tmpcstlast
```

```python
del tmpcstname['counts']
del tmpcstMid['counts']
del tmpcstlast['counts']
```

```python
tmpcstdf = pd.merge(tmpcstname, tmpcstMid, on=['CustomerID'])
cstdf = pd.merge(tmpcstdf, tmpcstlast, on=['CustomerID'])
```

# Creation of Star Schema

```python
# Import the package pymysql for making a connection to the database
import pymysql
```

```python
# Connect to the database
connection = pymysql.connect(host='localhost',
                         user='root',
                         password='Csoares#12',
                         db='dm_etl1')
```

```python
connection
```

```python
# create cursor
cursor=connection.cursor()
```

```python
cursor.execute("Create DATABASE newschema3")
connection.commit()
```

cursor.execute("USE newschema2")
connection.commit()

```python
# create the Customer Table
cursor.execute("""CREATE TABLE CustDetails
    (CustomerID INT UNIQUE,
     CustFirstName varchar(255) NULL,
     CustMiddleInitial varchar(255) NULL,
     CustLastName varchar(255) NULL,
     PRIMARY KEY (CustomerID)
     )
     """)

connection.commit()
```

```python
# create the Employee Table
cursor.execute("""CREATE TABLE EmpDetails
    (EmployeeID INT Unique,
     EmpFirstName varchar(255) NULL,
     EmpMiddleInitial varchar(255) NULL,
     EmpLastName varchar(255) NULL,
     PRIMARY KEY (EmployeeID)
     )
     """)

connection.commit()
```

```python
# create the Product Table
cursor.execute("""CREATE TABLE ProductDetails
    (ProductID INT Unique,
     ProductName varchar(255) NULL,
     Price float NULL,
     PRIMARY KEY (ProductID)
     )
     """)

connection.commit()
```

```python
# create the sales table
cursor.execute("""CREATE TABLE SalesDetails
    (SalesID int Unique,
     CustomerID int NULL,
     EmployeeID int NULL,
     ProductID int NULL,
     Quantity int NULL,
     TotalPrice float NULL,
     FOREIGN KEY (CustomerID) REFERENCES  CustDetails(CustomerID),
     FOREIGN KEY (EmployeeID) REFERENCES EmpDetails(EmployeeID),
     FOREIGN KEY (ProductID) REFERENCES ProductDetails(ProductID)
     )
     """)

connection.commit()
```

```python
# creating column list for insertion
cols = "`,`".join([str(i) for i in cstdf.columns.tolist()])

# Insert DataFrame recrds one by one.
for i,row in cstdf.iterrows():
    sql = "INSERT INTO `custdetails` (`" +cols + "`) VALUES (" + "%s,"*(len(row)-1) + "%s)"
    cursor.execute(sql, tuple(row))

    # the connection is not autocommitted by default, so we must commit to save our changes
    connection.commit()
```

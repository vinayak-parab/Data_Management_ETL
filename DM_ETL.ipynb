# Perform the required Transformations

```python
# calculate the frequency counts of each value in CustMiddleInitial column
df['CustMiddleInitial'].value_counts(dropna=False)
```

```python
df['CustMiddleInitial'].value_counts()['4']
```

```python
# replace the value 4 in CustMiddleInitial column with None
df.loc[df.CustMiddleInitial == "4", "CustMiddleInitial"] = None
```

```python
df['CustMiddleInitial'].value_counts(dropna=False)
```

```python
df.head()
```

```python
# calculate frequency of values in the Price Column
df['Price'].value_counts(dropna=False)
```

```python
# replace the $ and € symbol with a blank character in the Price column to maintain Consistency
df['Price'] = df['Price'].str.replace(r'[^\d.]+','')
```

```python
# check the frequency of Price Value to see if the currency characters have been removed
df['Price'].value_counts(dropna=False)
```

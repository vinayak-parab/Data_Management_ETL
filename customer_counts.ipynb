```python
# import the necessary packages
import pandas as pd
import mysql.connector
print("packages imported")
```

```python
# Set the values to display all the counts in one output cell
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.float_format = "{:,.2f}".format
```

```python
# make a connection to mysql database for extracting the raw data
mydb = mysql.connector.connect(host='localhost', database='salesdb', user='root', password='Csoares#12')
df = pd.read_sql('SELECT * FROM sales_data_etl', con=mydb)
```

```python
df
```

```python
# perform group by operation to get the counts 
# tmpcstdf = df.groupby(['CustomerID','CustFirstName','CustMiddleInitial','CustLastName'])['SalesID'].nunique()
tmpcstname = df.groupby(['CustomerID'])['CustFirstName'].value_counts(dropna = False)
tmpcstname=tmpcstname.to_frame()
tmpcstname.columns=['counts']
tmpcstname=tmpcstname.reset_index(level=['CustFirstName', 'CustomerID'])
tmpcstname=tmpcstname.drop_duplicates(subset='CustomerID', keep='first')
tmpcstname
```

```python
# perform group by operation to get the counts 
# tmpcstdf = df.groupby(['CustomerID','CustFirstName','CustMiddleInitial','CustLastName'])['SalesID'].nunique()
tmpcstMid = df.groupby(['CustomerID'])['CustMiddleInitial'].value_counts(dropna = False)
tmpcstMid=tmpcstMid.to_frame()
tmpcstMid.columns=['counts']
tmpcstMid=tmpcstMid.reset_index(level=['CustMiddleInitial', 'CustomerID'])
tmpcstMid=tmpcstMid.drop_duplicates(subset='CustomerID', keep='first')
tmpcstMid
```

```python
# perform group by operation to get the counts 
# tmpcstdf = df.groupby(['CustomerID','CustFirstName','CustMiddleInitial','CustLastName'])['SalesID'].nunique()
tmpcstlast = df.groupby(['CustomerID'])['CustLastName'].value_counts(dropna = False)
tmpcstlast=tmpcstlast.to_frame()
tmpcstlast.columns=['counts']
tmpcstlast=tmpcstlast.reset_index(level=['CustLastName', 'CustomerID'])
tmpcstlast=tmpcstlast.drop_duplicates(subset='CustomerID', keep='first')
tmpcstlast
```

```python
del tmpcstname['counts']
del tmpcstMid['counts']
del tmpcstlast['counts']
```

```python
tmpcstdf = pd.merge(tmpcstname, tmpcstMid, on=['CustomerID'])
cstdf = pd.merge(tmpcstdf, tmpcstlast, on=['CustomerID'])
```
